```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

```python
df=pd.read_csv(r"C:\Users\Dayanand\Desktop\Python\ML Projects\Python_Project_9_Ran.csv")
df
```

```python
df.isnull().sum()
```

```python
df=df.drop("company", axis=1)
df
```

```python
df=df.dropna(subset=["country"])
df.isnull().sum()
```

```python
df=df.dropna(subset=["children"])
```

```python
df["agent"]=df["agent"].fillna(df["agent"].mean())
df.isnull().sum()
```

```python
df
```

```python
df["reservation_status_day"]=pd.to_datetime(df.reservation_status_date, format="%d/%m/%Y").dt.day
df["reservation_status_month"]=pd.to_datetime(df.reservation_status_date, format="%d/%m/%Y").dt.month
df
```

```python
df=df.drop("reservation_status_date", axis=1)
df
```

```python
df.loc[:, "adults":"deposit_type"]
```

```python
df["hotel"]=LabelEncoder().fit_transform(df["hotel"])
df["arrival_date_month"]=LabelEncoder().fit_transform(df["arrival_date_month"])
df["meal"]=LabelEncoder().fit_transform(df["meal"])
df["country"]=LabelEncoder().fit_transform(df["country"])
df["market_segment"]=LabelEncoder().fit_transform(df["market_segment"])
df["distribution_channel"]=LabelEncoder().fit_transform(df["distribution_channel"])
df["reserved_room_type"]=LabelEncoder().fit_transform(df["reserved_room_type"])
df["assigned_room_type"]=LabelEncoder().fit_transform(df["assigned_room_type"])
df["deposit_type"]=LabelEncoder().fit_transform(df["deposit_type"])
df["customer_type"]=LabelEncoder().fit_transform(df["customer_type"])
df["reservation_status"]=LabelEncoder().fit_transform(df["reservation_status"])
df
```

```python
df.corr()
```

```python
x=df.drop(["arrival_date_month","stays_in_weekend_nights","children", "is_canceled"], axis=1)
x
```

```python
y=df["is_canceled"]
y
```

```python
cols=x.columns
cols
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2)
```

```python
x_train.shape, x_test.shape
```

```python
x_train=StandardScaler().fit_transform(x_train)
x_train=pd.DataFrame(x_train, columns=[cols])
x_train
```

```python
x_test=StandardScaler().fit_transform(x_test)
x_test=pd.DataFrame(x_test, columns=[cols])
x_test
```

```python
model=RandomForestClassifier(n_estimators=500, criterion='entropy', max_depth=12,min_samples_split=5,min_samples_leaf=2)
model.fit(x_train, y_train)
model.score(x_train, y_train), model.score(x_test, y_test)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
print("Accuracy Score",accuracy_score(y_test, y_pred))
print("Precision Score",precision_score(y_test, y_pred))
print("Recall Score",recall_score(y_test, y_pred))
print("F1 Score",f1_score(y_test, y_pred))
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
