```python
x=df.drop("Quality_Category", axis=1)
x
```

```python
y=df["Quality_Category"]
y
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2)
```

```python
x_train.shape, x_test.shape
```

```python
cols=x.columns
cols
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
x_train=RobustScaler().fit_transform(x_train)
```

```python
x_train=pd.DataFrame(x_train, columns= [cols])
x_train
```

```python
x_test=RobustScaler().fit_transform(x_test)
x_test=pd.DataFrame(x_test, columns= [cols])
x_test
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
models=[LogisticRegression, KNeighborsClassifier, SVC, DecisionTreeClassifier, RandomForestClassifier, GaussianNB]
for mod in models:
    model=mod()
    model.fit(x_train, y_train)
    print(mod, model.score(x_train, y_train), model.score(x_test, y_test))
```

```python
#Parameter Tuning with SVC
```

```python
for i in range(1,100,10):
    model=SVC(C=i)
    model.fit(x_train, y_train)
    print(i, model.score(x_train, y_train), model.score(x_test, y_test))
```

```python
model = SVC(C=31)
model.fit(x_train, y_train)
model.score(x_train, y_train), model.score(x_test, y_test)
```
